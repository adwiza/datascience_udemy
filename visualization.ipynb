# Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
```

```python
fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])
ax.plot(x, y, c='red', alpha=.5, marker='^', lw=2.5, label='Sine')
ax.plot(x, z, c='green', alpha=.5, marker='x', label='Cosine') # transparency

ax.set_xlim(0, 10)
ax.set_ylim(-1, 1)
ax.legend()

ax.set_xlabel('Time')
ax.set_ylabel('Amp')
ax.set_title('My graph in jupyter notebook')
```

# Subplots

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 7))
ax[0].plot(x, y, c='red')
ax[1].plot(x, z, c='green')
```

# Scatter and Line Plot

```python
import plotly.graph_objs as go
import numpy as np
```

```python
x = np.random.randint(1, 50, 10)
y = np.random.randint(1, 20, 10)
```

```python
data = [go.Scatter(x=x, y=y, mode='markers')]
```

```python
pyo.plot(data, filename='scatter-plot.html')
```

```python
x = np.arange(1, 20, 1)
y = np.random.randint(1, 20, 20)
```

```python
data = [go.Scatter(x=x, y=y, mode='lines+markers')]
```

```python
pyo.plot(data, filename='lines-plot.html')
```

```python
x = np.arange(1, 20, 1)
y = np.random.randint(1, 20, 20)
```

```python
t1 = go.Scatter(x=x, y=y, mode='lines+markers')
t2 = go.Scatter(x=x, y=y+3, mode='lines')
```

```python
data = [t1, t2]
```

```python
pyo.plot(data, filename='2lines-plot.html')
```

# Bar chart

```python
data = [go.Bar(
    x = ['Grade - 1', 'Grade - 2', 'Grade - 3'],
    y = [23, 45, 37]
)]
```

```python
 pyo.plot(data, filename='bar.html')
```

# Bubble chart

```python
x = np.arange(1, 20, 1)
y = np.random.randint(1, 20, 20)
z = np.random.randint(1, 100, 20)
```

```python
data = [go.Scatter(x=x, y=y, mode='markers', marker=dict(size=z))]
```

```python
pyo.plot(data, filename='bubble.html')
```

# Histogram and distribution plot

```python
x = np.random.randint(1, 100, 500)
```

```python
data = [go.Histogram(
    x = x,
    xbins = dict(start=1, end=100, size=5)
)]
```

```python
pyo.plot(data, filename='histogram.html')
```
